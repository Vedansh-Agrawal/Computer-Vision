## Video to Images

```python
import cv2
import os
import matplotlib.pyplot as plt

video_path = 'Chroma_key.mp4'
image_path = 'images'

video = cv2.VideoCapture(video_path)
os.makedirs(image_path, exist_ok=True)

frame_count=0
while True:
    ret, frame = video.read()
    if not ret:
        break

    image_paths = os.path.join(image_path, f"Chroma_{frame_count:04d}.png")
    cv2.imwrite(image_paths, frame)

    frame_count += 1
    
video.release()
```

## Images to Videos

```python
import cv2
import os
import matplotlib.pyplot as plt

output_video = 'plane.mp4'
frame_rate = 30

images = sorted([img for img in os.listdir(image_path)])

first_img = cv2.imread(os.path.join(image_path, images[0]))
h, w, _ = first_img.shape

video_writer = cv2.VideoWriter(output_video, cv2.VideoWriter_fourcc(*'mp4v'), frame_rate, (w, h))

for image in images:
    frame = cv2.imread(os.path.join(image_path, image))
    print(type(frame), w, h)
    video_writer.write(frame)

video_writer.release()
```

## Capturing videos from camera

```python
import cv2
import os
import matplotlib.pyplot as plt


camera_images = 'camera_images'
os.makedirs(camera_images, exist_ok=True)

camera = cv2.VideoCapture(0)

cv2.namedWindow('Video', cv2.WINDOW_AUTOSIZE)

frame_count = 0
while True:
    ret, frame = camera.read()
    
    cv2.imshow('Video', frame)

    image_paths = os.path.join(camera_images, f"Frame_{frame_count:04d}.png")
    cv2.imwrite(image_paths, frame)
    frame_count += 1

    if cv2.waitKey(1) == 27:
        break
    
camera.release()
cv2.destroyAllWindows()
```

## Chroma Keying

```python
import cv2
import os
import matplotlib.pyplot as plt

background = cv2.VideoCapture('Background.mp4')
foreground = cv2.VideoCapture('Foreground.mp4')


green_lr = (0, 170, 0)
green_up = (55, 255, 55)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video = cv2.VideoWriter('Chroma_key.mp4', fourcc, 24, (int(background.get(3)), int(background.get(4))))



while True:

    back_ret, back_frames = background.read()
    fore_ret, fore_frames = foreground.read()


    if not back_ret or not fore_ret:
        break

    fore_frames = cv2.resize(fore_frames, (back_frames.shape[1], back_frames.shape[0]))

    mask = cv2.inRange(fore_frames, green_lr, green_up)
    inverted_mask = cv2.bitwise_not(mask)


    composite_frame = cv2.bitwise_and(back_frames, back_frames, mask=mask)

    composite_frame += cv2.bitwise_and(fore_frames, fore_frames, mask=inverted_mask)


    video.write(composite_frame)

background.release()
foreground.release()
video.release()

cv2.destroyAllWindows()
```
